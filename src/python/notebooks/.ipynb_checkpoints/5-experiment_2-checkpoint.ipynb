## 5. Function to plot two or more timeseries

```python
def plot_values(timestamp, *args):
    import matplotlib

    if len(args) % 2 != 0:
        raise ValueError("Every feature should have a corresponding name")

    # Use a predefined style
    plt.style.use('ggplot')

    # Set figure size
    fig, ax = plt.subplots(figsize=(12,6))

    # Create a colormap that will generate colors
    colormap = matplotlib.colormaps['tab10']

    # Plot data with customized line for each feature
    for i in range(0, len(args), 2):
        feature_name = args[i]
        feature_values = args[i+1]
        color = colormap(i // 2 / (len(args)//2))
        ax.plot(timestamp, feature_values, label=feature_name, color=color, linewidth=1)

    # Set labels with improved readability
    ax.set_xlabel('Date')
    ax.yaxis.set_label_position("right") # This line moves y label to the right
    ax.set_ylabel('Feature Values')
    ax.yaxis.tick_right() # This line moves y axis to the right
    ax.set_title('Features over Time')

    # Format the timestamps and set locator
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) # format as year-month-day
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3)) # display a label at the start of every 3rd month

    # Rotate x-axis labels for better visibility
    plt.xticks(rotation=45)

    # Add a grid
    ax.grid(True)

    # Add legend
    ax.legend()

    # Show plot
    plt.show()
```

## 6. Function to scale the data and split it for training and testing

```python
def data_preperation(df, lookback, future, scale, test_size=0.1):
    # Convert 'Date' column to datetime
    try:
        date_train = pd.to_datetime(df['timestamp'])
        df = df.drop(columns=['timestamp'])
    except:
        date_train = [0]


    # Ensure all data is float type
    df = df.astype(float)

    # Split data into training and testing before scaling
    df_train, df_test = train_test_split(df, test_size=test_size, shuffle=False)

    # Scale data
    df_train_scaled = scale.fit_transform(df_train)
    df_test_scaled = scale.transform(df_test)  # use the scaler fitted on the training data

    # Create the feature and target arrays
    X_train, y_train = [], []
    for i in range(lookback, len(df_train_scaled)-future+1):
        X_train.append(df_train_scaled[i-lookback:i, :])
        y_train.append(df_train_scaled[i+future-1:i+future, 0])

    X_test, y_test = [], []
    for i in range(lookback, len(df_test_scaled)-future+1):
        X_test.append(df_test_scaled[i-lookback:i, :])
        y_test.append(df_test_scaled[i+future-1:i+future, 0])

    # Convert to numpy arrays
    X_train = np.array(X_train)
    y_train = np.array(y_train)
    X_test = np.array(X_test)
    y_test = np.array(y_test)

    return X_train, X_test, y_train, y_test, date_train
```

## 7. Function that defines the BiRNN hypertuned on ta_df dataset

```python
def Oldta_df_birnn(X, y):
    model_name="rnn_1_"+str(int(time.time()))
    model=Sequential(name=model_name)

    model.add(Bidirectional(SimpleRNN(units=448, return_sequences=True), input_shape=(X.shape[1], X.shape[2])))
    model.add(Dropout(0.2))
    model.add(Bidirectional(SimpleRNN(units=256, return_sequences=False)))
    model.add(Dense(25))
    model.add(Dense(1))

    adam = optimizers.Adam(learning_rate=0.001)
    model.compile(loss='huber_loss',optimizer=adam)

    model.fit(X, y, validation_split=0.1, epochs=50, batch_size=32)
    return model
```

```python
def ta_df_birnn(X, y):
    model_name="rnn_1_"+str(int(time.time()))
    model=Sequential(name=model_name)

    model.add(Bidirectional(SimpleRNN(units=512, return_sequences=True), input_shape=(X.shape[1], X.shape[2])))
    model.add(Dropout(0.2))
    model.add(Bidirectional(SimpleRNN(units=256, return_sequences=False)))
    model.add(Dense(90))
    model.add(Dense(1))

    adam = optimizers.Adam(learning_rate=0.001)
    model.compile(loss='huber_loss',optimizer=adam)

    model.fit(X, y, validation_split=0.1, epochs=50, batch_size=32)
    return model
```

## 9. Function that defines the BiLSTM hypertuned on ta_df dataset

```python
def ta_df_bilstm(X, y):
    model_name="lstm_1_"+str(int(time.time()))
    model=Sequential(name=model_name)

    model.add(Bidirectional(LSTM(units=480, return_sequences=True), input_shape=(X.shape[1], X.shape[2])))
    model.add(Dropout(0.2))
    model.add(Bidirectional(LSTM(units=96, return_sequences=False)))
    model.add(Dense(10))
    model.add(Dense(1))

    adam = optimizers.Adam(learning_rate=0.001)
    model.compile(loss="mean_squared_error",optimizer=adam)

    model.fit(X, y, validation_split=0.1, epochs=50, batch_size=32)
    return model
```

```python
def oc_ta_df_bilstm(X, y):
    model_name="lstm_1_"+str(int(time.time()))
    model=Sequential(name=model_name)

    model.add(Bidirectional(LSTM(units=352, return_sequences=True), input_shape=(X.shape[1], X.shape[2])))
    model.add(Dropout(0.2))
    model.add(Bidirectional(LSTM(units=32, return_sequences=False)))
    model.add(Dense(25))
    model.add(Dense(1))

    adam = optimizers.Adam(learning_rate=0.001)
    model.compile(loss="huber_loss",optimizer=adam)

    model.fit(X, y, validation_split=0.1, epochs=50, batch_size=32)
    return model
```

## 11. Function for selecting the correct model, training it, making predictions and then evaluating them.

```python
import os
from datetime import datetime
from google.colab import drive
drive.mount('/content/drive')

def predict_price_model(df, model_type):
    scale = MinMaxScaler(feature_range=(0,1))
    X_train, X_test, y_train, y_test, date_train = data_preperation(df, 60, 1, scale)

    if model_type == "ta_df_birnn":
        model = ta_df_birnn(X_train, y_train)
    elif model_type == "oc_ta_df_birnn":
        model = oc_ta_df_birnn(X_train, y_train)
    elif model_type == "ta_df_bilstm":
        model = ta_df_bilstm(X_train, y_train)
    elif model_type == "oc_ta_df_bilstm":
        model = oc_ta_df_bilstm(X_train, y_train)
    else:
        print(f"Unknown model type: {model_type}")
        return

    predictions = model.predict(X_test)

    price_scaler = MinMaxScaler(feature_range=(0,1))
    price_scaler.min_, price_scaler.scale_ = scale.min_[0], scale.scale_[0]

    # Now inverse_transform the predictions
    predictions = price_scaler.inverse_transform(predictions)
    y_test = y_test.reshape(-1, 1)  # reshaping
    y_test = price_scaler.inverse_transform(y_test)

    # Mean Absolute Error (MAE)
    mae = mean_absolute_error(y_test, predictions)
    print("Mean Absolute Error (MAE):", mae)

    # Mean Squared Error (MSE)
    mse = mean_squared_error(y_test, predictions)
    print("Mean Squared Error (MSE):", mse)

    # Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    print("Root Mean Squared Error (RMSE):", rmse)

    # Mean Absolute Percentage Error (MAPE)
    mape = np.mean(np.abs((y_test - predictions) / y_test)) * 100
    print("Mean Absolute Percentage Error (MAPE):", mape)

    # R-squared (R2)
    r2 = r2_score(y_test, predictions)
    print("R-squared (R2):", r2)

    # Mean Directional Accuracy (MDA)
    y_test_shifted = np.roll(y_test, shift=-1)[:-1]  # Shift y_test by one period
    predictions_shifted = np.roll(predictions, shift=-1)[:-1]  # Shift predictions by one period
    correct_directions = np.sign(predictions_shifted - y_test[:-1]) == np.sign(y_test_shifted - y_test[:-1])
    mda = np.mean(correct_directions)

    print("Mean Directional Accuracy (MDA):", mda)

    plot_values(range(len(y_test)), 'test', y_test, 'predictions', predictions)
    now = datetime.now()  # get current date and time
    timestamp = now.strftime("%Y%m%d_%H%M%S")  # format date and time as a string

    model_directory = "/content/drive/MyDrive/MyModels"  # specify the directory in Google Drive to save the models
    os.makedirs(model_directory, exist_ok=True)  # create the directory if it doesn't exist

    model_name = f"{model_type}_{timestamp}.h5"  # append the timestamp to the model name
    model_path = os.path.join(model_directory, model_name)  # full path to save the model

    model.save(model_path)
    print(f"Model saved as {model_path}")
    # Return a dictionary of performance metrics
    return {"MAE": mae, "MSE": mse, "RMSE": rmse, "MAPE": mape, "R2": r2, "MDA": mda}



```

```python
def run_experiments(dataframes, model_names, num_runs):
    # Dictionary to hold results
    results = {}

    # Loop over all combinations of dataframes and model names
    for df_name, df in dataframes.items():
        for model_name in model_names:
            # List to hold results for this combination of df and model
            df_model_results = []

            for i in range(num_runs):
                # Run model and get some measure of performance
                performance = predict_price_model(df, model_name)

                # Append performance to results
                df_model_results.append(performance)

            # Store list of results in dictionary
            results[(df_name, model_name)] = df_model_results

    return results
```

```python
def print_experiment_results(results):
    # Initialize an empty list to hold rows of the dataframe
    df_rows = []

    # Loop over each combination of dataframe and model
    for (df_name, model_name), perf_list in results.items():
        # Loop over each run for this combination
        for i, performance in enumerate(perf_list):
            # Add a row to the dataframe for this run
            df_rows.append({
                'DataFrame': df_name,
                'Model': model_name,
                'Run': i+1,
                'MAE': performance['MAE'],
                'MSE': performance['MSE'],
                'RMSE': performance['RMSE'],
                'MAPE': performance['MAPE'],
                'R2': performance['R2'],
                'MDA': performance['MDA']
            })

    # Create the dataframe
    results_df = pd.DataFrame(df_rows)

    # Compute the average performance for each combination of DataFrame and Model
    average_results = results_df.groupby(['DataFrame', 'Model']).mean().reset_index()

    # Return the average results
    return results_df, average_results
```

## 12. Create 5 models for each dataframe and each model type then evaluate them and print the results and the averaged results

```python
dataframes = {"ta_df": ta_df, "oc_ta_df": oc_ta_df}
model_names = ["ta_df_birnn", "ta_df_bilstm"]
results = run_experiments(dataframes, model_names, 5)
results_df, average_results = print_experiment_results(results)
print(results_df)
print(average_results)
```
